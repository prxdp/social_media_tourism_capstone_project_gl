#### Outlier Treatment

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
Cont=['Yearly_avg_view_on_travel_page', 'total_likes_on_outstation_checkin_given','yearly_avg_Outstation_checkins','Yearly_avg_comment_on_travel_page','total_likes_on_outofstation_checkin_received','montly_avg_comment_on_company_page','Daily_Avg_mins_spend_on_traveling_page']
plt.figure(figsize=(10,10))
df_imputed[Cont].boxplot(vert=0)
plt.title('With Outliers',fontsize=16)
plt.show()
```

```python
for column in df_imputed[Cont].columns:
    lr,ur=remove_outlier(df_imputed[column])
    df_imputed[column]=np.where(df_imputed[column]>ur,ur,df_imputed[column])
    df_imputed[column]=np.where(df_imputed[column]<lr,lr,df_imputed[column])
```

Creating a function to evaluate various model matrices

```python
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name, data_name, results_df):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Calculate the confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Calculate RMSE
    rmse = mean_squared_error(y_test, y_pred, squared=False)

    # Calculate R-squared
    r2 = r2_score(y_test, y_pred)

    # Calculate ROC and AUC
    if len(set(y_test)) == 2:
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        auc = roc_auc_score(y_test, y_prob)
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(auc))
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {model_name} - {data_name}')
        plt.legend(loc='lower right')
        plt.show()

    # Calculate classification report
    class_report = classification_report(y_test, y_pred)

    # Calculate test and train model accuracy
    train_acc = model.score(X_train, y_train)
    test_acc = model.score(X_test, y_test)

    # Add results to the dataframe
    results_df = results_df.append({
        'Model': model_name,
        'Data': data_name,
        'Confusion Matrix': cm,
        'RMSE': rmse,
        'R-squared': r2,
        'AUC': auc if len(set(y_test)) == 2 else None,
        'Train Accuracy': train_acc,
        'Test Accuracy': test_acc
    }, ignore_index=True)

    print(f'Evaluation for {model_name} - {data_name}:')
    print("Confusion Matrix:")
    print(cm)
    print("RMSE:", rmse)
    if r2 is not None:
        print("R-squared:", r2)
    if len(set(y_test)) == 2:
        print("AUC:", auc)
    print("Classification Report:")
    print(class_report)
    print("Train Accuracy:", train_acc)
    print("Test Accuracy:", test_acc)
    print()

    return results_df
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold


results_df = pd.DataFrame(columns=['Model', 'Data', 'Accuracy Range'])

# Define the number of cross-validation folds
n_folds = 5

# Define the cross-validation strategy (StratifiedKFold is used here for classification tasks)
cv = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

# Evaluate models for 'mobile_data' and 'laptop_data' separately
for data_name, X_train, y_train in [('mobile_data', X_train_mobile, y_train_mobile), ('laptop_data', X_train_laptop, y_train_laptop)]:
    for model_name, model in models.items():
        # Perform cross-validation
        scores = cross_val_score(model, X_train, y_train, cv=cv)

        # Calculate the accuracy range (minimum and maximum accuracy)
        accuracy_range = (scores.min().round(2), scores.max().round(2))

        # Add results to the dataframe
        results_df = results_df.append({
            'Model': model_name,
            'Data': data_name,
            'Accuracy Range': accuracy_range
        }, ignore_index=True)

# Display the results dataframe
print("Results:")
results_df
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Define a function to evaluate the model and generate various metrics
def evaluate_model(model, data, features, target, data_name):
    X = data[features]
    y = data[target]

    # Split the data into train and test sets (you can adjust the ratio as needed)
    train_size = int(0.7 * len(X))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Bagging with Decision Tree classifier
    bagging_model = BaggingClassifier(base_estimator=model, n_estimators=100)
    bagging_model.fit(X_train, y_train)

    # Make predictions
    y_pred = bagging_model.predict(X_test)

    # Calculate metrics
    classification = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    auc_score = roc_auc_score(y_test, bagging_model.predict_proba(X_test)[:,1])
    fpr, tpr, _ = roc_curve(y_test, bagging_model.predict_proba(X_test)[:,1])
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    accuracy = np.mean(y_pred == y_test)

    # Plot ROC curve
    auc = roc_auc_score(y_test, y_pred)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {data_name}')
    plt.legend(loc='lower right')
    plt.show()

    print(f"Classification Report ({data_name}):\n{classification}")
    print(f"Confusion Matrix ({data_name}):\n{conf_matrix}")
    print(f"AUC Score ({data_name}): {auc_score:.4f}")
    print(f"RMSE ({data_name}): {rmse:.4f}")
    print(f"Accuracy ({data_name}): {accuracy:.4f}")

# Use the evaluate_model function for both mobile_data and laptop_data
evaluate_model(DecisionTreeClassifier(), mobile_data, features, target, 'mobile_data')
evaluate_model(DecisionTreeClassifier(), laptop_data, features, target, 'laptop_data')
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV

# Define a function to tune and evaluate the model
def tune_and_evaluate_model(model, param_grid, data, features, target, data_name):
    X = data[features]
    y = data[target]

    # Split the data into train and test sets (you can adjust the ratio as needed)
    train_size = int(0.7 * len(X))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Hyperparameter tuning using GridSearchCV
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='roc_auc')
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_

    # Make predictions
    y_pred = best_model.predict(X_test)

    # Calculate metrics
    classification = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    auc_score = roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1])
    fpr, tpr, _ = roc_curve(y_test, best_model.predict_proba(X_test)[:,1])
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    accuracy = np.mean(y_pred == y_test)
    auc = roc_auc_score(y_test, y_pred)

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {data_name}')
    plt.legend(loc='lower right')
    plt.show()

    print(f"Classification Report ({data_name}):\n{classification}")
    print(f"Confusion Matrix ({data_name}):\n{conf_matrix}")
    print(f"AUC Score ({data_name}): {auc_score:.4f}")
    print(f"RMSE ({data_name}): {rmse:.4f}")
    print(f"Accuracy ({data_name}): {accuracy:.4f}")

# Define the hyperparameter grid for tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0]
}

# Use the tune_and_evaluate_model function for both mobile_data and laptop_data
tune_and_evaluate_model(AdaBoostClassifier(base_estimator=DecisionTreeClassifier()), param_grid, mobile_data, features, target, 'mobile_data')
tune_and_evaluate_model(AdaBoostClassifier(base_estimator=DecisionTreeClassifier()), param_grid, laptop_data, features, target, 'laptop_data')
```
